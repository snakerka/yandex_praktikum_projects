**Определение выгодного тарифа для телеком компании**

- минута разговора: 3 рубля
- сообщение: 3 рубля
- 1 Гб интернет-трафика: 200 рублей

**Тариф «Ультра»**
- Ежемесячная плата: 1950 рублей
- Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
- Стоимость услуг сверх тарифного пакета:
- минута разговора: 1 рубль
- сообщение: 1 рубль
- 1 Гб интернет-трафика: 150 рублей

#### Посчитаем для каждого пользователя:
- количество сделанных звонков и израсходованных минут разговора по месяцам;
- количество отправленных сообщений по месяцам;
- объем израсходованного интернет-трафика по месяцам;
- помесячную выручку с каждого пользователя (вычтем бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; остаток умножим на значение из тарифного плана; прибавим абонентскую плату, соответствующую тарифному плану).

```python
# количество сделанных звонков и израсходованных минут разговора по месяцам

df_calls['month'] = df_calls['call_date'].dt.month # месяц вызова
df_calls_duration_per_month = df_calls.pivot_table(
    index = ['user_id','month'],
    values = 'duration',
    aggfunc = ['count','sum'])
df_calls_duration_per_month.columns = ['total_call_number','total_call_duration']
df_calls_duration_per_month
```

```python
# количество отправленных сообщений по месяцам

df_messages['month'] = df_messages['message_date'].dt.month # месяц вызова

df_messages_per_month = df_messages.pivot_table(
    index = ['user_id','month'],
    values = 'message_date',
    aggfunc = 'count')
df_messages_per_month.columns = ['total_messages']
df_messages_per_month
```

```python
# объем израсходованного интернет-трафика по месяцам

df_internet['month'] = df_internet['session_date'].dt.month # месяц вызова

df_session_per_month = df_internet.pivot_table(
    index = ['user_id','month'],
    values = 'mb_used',
    aggfunc = 'sum')
df_session_per_month.columns = ['total_session']
df_session_per_month
```

```python
# создаем таблицу выручки по каждому клиенту и его активностей в разрезе  месяца

revenue_df_list = [df_messages_per_month, df_session_per_month]
revenue = df_calls_duration_per_month
for df_list in revenue_df_list:
    revenue = revenue.join(df_list, on = ['user_id', 'month'])

revenue = revenue.reset_index()
revenue.info()
```

```python
# Заменим пустые данные по сообщениям и интернет трафику на 0, 
# так как данные пропуски означают, что клиент данной услугой не пользовался

revenue['total_messages'] = revenue['total_messages'].fillna(value = 0)
revenue['total_session'] = revenue['total_session'].fillna(value = 0)

revenue.info()
```

```python
# добавим информацию о названии тарифного плана

name_tariff = pd.DataFrame(df_users, columns =['user_id','tariff']) # создаем новый DataFrame с названием тарифа
name_tariff['user_id'] = name_tariff['user_id'].astype('int')
name_tariff.info()

revenue = revenue.merge(name_tariff, on = 'user_id')
revenue.head()
```

```python
# переименуем столбцы для последующего merge

df_tariffs.columns = ['messages_included','mb_per_month_included',
                      'minutes_included','rub_monthly_fee','rub_per_gb',
                      'rub_per_message','rub_per_minute', 'tariff'] 
df_tariffs
```

```python
# Добавим информацию по условиям тарифов

revenue = revenue.merge(df_tariffs, on = 'tariff')
revenue.head()
```

```python
# Переведем в ГБ значения трафика и округлим до большего значения

revenue['total_session_gb'] = np.ceil(revenue['total_session'])/1024
revenue['gb_per_month_included'] = revenue['mb_per_month_included'] / 1024
revenue.head()
```

```python
# Добавим столбцы с перелимитами по тарифам
# Перелемит минут

revenue['leftover_calls'] = revenue.apply(
    lambda row: max(row['total_call_duration'] - row['minutes_included'],0), axis = 1
)

# Столбец с перелимитом сообщений

revenue['leftover_messages'] = revenue.apply(
    lambda row: max(row['total_messages'] - row['messages_included'],0), axis = 1
)

# Столбец с перелимитом ГБ

revenue['leftover_gb'] = revenue.apply(
    lambda row: max(row['total_session_gb'] - row['gb_per_month_included'],0), axis = 1
)
revenue.sample(10)
```

```python
# Стоимость использование услуг компании каждого клиента в разрезе месяцев

def clients_cost(row):
    cost_calls = row['leftover_calls'] * row['rub_per_minute']
    cost_messages = row['rub_per_message'] * row['leftover_messages']
    cost_internet = row['rub_per_gb'] * row['leftover_gb']
    total_cost = row['rub_monthly_fee'] + cost_messages + cost_calls + cost_internet
    return total_cost
revenue['cost'] = revenue.apply(clients_cost, axis = 1)
revenue.sample(10)
```

### <a id="step4"> Шаг 4. Проверим гипотезы </a>

**Гипотеза 1.** Средняя выручка пользователей тарифов Ultra и Smart различается

- Нулевая гипотеза - средняя выручка пользователей тарифов «Ультра» и «Смарт» равны, т.е. средние значения равны.
- Альтернативная гипотеза - средняя выручка пользователей тарифов «Ультра» и «Смарт» не равны, т.е. средние значения не равны.

Для проверки гипотезы будем использовать критерий p-value.

Критический уровень статистической значимости - 5 процентов

#### Выбор использования t-распределением Стьюдента для проверки гипотез обусловлен тем, что количество элементов выборки более 30

#### Нулевые гипотезы формируются на основании предположения, что величины равны

```python
sample_1 = revenue[revenue['tariff'] == 'ultra']['cost'] 
sample_2 = revenue[revenue['tariff'] == 'smart']['cost'] 

results = st.ttest_ind(
    sample_1, 
    sample_2)
alpha = 0.05

print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

#### Результат: Полученное значение p-value очень мало, что позволяет сделать вывод, что средние значения для тарифов не равны.

**Гипотеза 2.** Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов;

- Нулевая гипотеза - cредняя выручка пользователей из Москвы равна выручке пользователей из других регионов.
- Альтернативная гипотеза - cредняя выручка пользователей из Москвы не равна выручке пользователей из других регионов..

Для проверки гипотезы будем использовать критерий p-value.

Критический уровень статистической значимости - 5 процентов

```python
# Добавим в таблицу информацию о городах

df_users_city = df_users[['user_id', 'city']] # выбираем нужны столбцы

revenue = revenue.merge(df_users_city, on = 'user_id') # добавляем информацию о городах
```

```python
sample_1 = revenue[revenue['city'] == 'Москва']['cost'] 
sample_2 = revenue[revenue['tariff'] != 'Москва']['cost'] 

results = st.ttest_ind(
    sample_1, 
    sample_2)
alpha = 0.05

print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

#### Результат: Полученное значение p-value позволяет сделать вывод, что средние значения выручки для пользователей из Москвы и пользователей из других регионов равны.
