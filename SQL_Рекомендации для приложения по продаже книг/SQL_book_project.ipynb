- Определим автора с самой высокой средней оценкой книг — учитываем только книги с 50 и более оценками;
- Посчитаем среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
# Для каждой книги посчитаем количество обзоров и среднюю оценку

query = """ WITH avg_rating as 
(SELECT ratings.book_id, AVG(ratings.rating) as avg_rating FROM ratings GROUP BY ratings.book_id), 
cnt_review as (SELECT reviews.book_id, COUNT(reviews.review_id) as cnt_review FROM reviews GROUP BY reviews.book_id)
SELECT title, avg_rating, cnt_review
FROM books
JOIN avg_rating ON books.book_id = avg_rating.book_id
JOIN cnt_review ON books.book_id = cnt_review.book_id
ORDER BY avg_rating DESC, cnt_review DESC
"""
pd.io.sql.read_sql(query,con = engine)
```

Самая популярная книга - A Dirty Job (Grim Reaper #1).

```python
# Определим издательство, которое выпустило наибольшее число книг толще 50 страниц

query = """SELECT publisher, COUNT(book_id) as cnt_book
FROM publishers
JOIN books on books.publisher_id = publishers.publisher_id
WHERE books.num_pages > 50
GROUP BY publisher
ORDER BY cnt_book DESC
LIMIT 1"""
pd.io.sql.read_sql(query, con = engine)
```

Больше всего книг выпустило издательство `Penguin Books`.

```python
# Определим автора с самой высокой средней оценкой книг — учитываем только книги с 50 и более оценками

query = """WITH avg_rating as 
    (SELECT ratings.book_id, AVG(ratings.rating) as avg_rating_book, COUNT(ratings.rating) as cnt_rating_book  
    FROM ratings 
    GROUP BY ratings.book_id)
SELECT author, COUNT(cnt_rating_book) as cnt_rating_author, AVG(avg_rating_book) as avg_rating_author
FROM books
JOIN avg_rating ON books.book_id = avg_rating.book_id
JOIN authors on books.author_id = authors.author_id
WHERE cnt_rating_book >=50
GROUP BY author
ORDER BY avg_rating_author DESC
"""
pd.io.sql.read_sql(query, con = engine)
```

Самая популярная писательница по оценкам пользователей, автор книг о Гарри Поттере - `Дж. Роулинг`.

```python
# Посчитаем среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

query = """WITH 
    cnt_review as (SELECT reviews.username, COUNT(reviews.review_id) as cnt_reviews
    FROM reviews
    GROUP BY reviews.username), 
    
    cnt_rating as (SELECT ratings.username, COUNT(ratings.rating) as cnt_rating
    FROM ratings
    GROUP BY ratings.username
    HAVING COUNT(ratings.rating) >50)
    
SELECT AVG(cnt_review.cnt_reviews)
FROM cnt_rating
JOIN cnt_review ON cnt_review.username = cnt_rating.username"""
pd.io.sql.read_sql(query, con = engine)
```

Среднее количество обзоров от пользователей с высоким количеством оценок - 24.
